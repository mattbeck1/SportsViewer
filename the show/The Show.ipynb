```python
import cv2
import numpy as np
import pytesseract
import matplotlib.pyplot as plt

pytesseract.pytesseract.tesseract_cmd = 'C:\Program Files\Tesseract-OCR\\tesseract.exe'  # your path may be different
```

```python
def get_scoreboard(name):
    img = cv2.imread(name)

    Y, X = 1760, 2770
    H, W = 290, 970
    scoreboard = img[Y:Y+H, X:X+W]
    gray_scoreboard = cv2.cvtColor(scoreboard, cv2.COLOR_BGR2GRAY)   
    scores = gray_scoreboard[110:, 180:490]

    return scores, gray_scoreboard

def get_homerun(name):
    img = cv2.imread(name)

    Y, X = 1620, 2890
    H, W = 140, 730
    home_run = img[Y:Y+H, X:X+W]
    grey_hr = cv2.cvtColor(home_run, cv2.COLOR_BGR2GRAY)   

    return grey_hr
```

```python
scores, gray_scoreboard = get_scoreboard('The Show Test 1.png')

plt.imshow(gray_scoreboard)
plt.show()
plt.imshow(scores)
plt.show()
```

```python
score1 = scores[45:140, :150]
score1_text = pytesseract.image_to_string(score1, lang='eng', config='--psm 13 --oem 3 -c tessedit_char_whitelist=0123456789')
print(score1_text.strip())
plt.imshow(score1)
plt.show()
```

```python
score2 = scores[45:140, 150:]
score2_text = pytesseract.image_to_string(score2, lang='eng', config='--psm 6 --oem 3 -c tessedit_char_whitelist=0123456789')
print(score2_text.strip())
plt.imshow(score2)
plt.show()
```
